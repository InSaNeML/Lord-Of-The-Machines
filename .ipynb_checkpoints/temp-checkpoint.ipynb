```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("./data/train_mini.csv", parse_dates=['send_date'])
df.head()
```

```python
df.info()
```

```python
df["month"] = df.send_date.map(lambda x: x.strftime('%m'))
df["date"] = df.send_date.map(lambda x: x.strftime('%d'))
df["hour"] = df.send_date.map(lambda x: x.strftime('%H'))
df["min"] = df.send_date.map(lambda x: x.strftime('%M'))
df.drop(["send_date"], axis=1, inplace=True)
df.head()
```

```python
daf = pd.read_csv("./data/campaign_data.csv")
daf.head()
```

```python
from sklearn.preprocessing import LabelEncoder

le_comm = LabelEncoder()
le_body = LabelEncoder()
le_sub = LabelEncoder()
le_url = LabelEncoder()
```

```python
le_comm.fit_transform(daf.communication_type)
le_body.fit_transform(daf.email_body)
le_sub.fit_transform(daf.subject)
le_url.fit_transform(daf.email_url)

daf.communication_type = le_comm.transform(daf.communication_type)
daf.email_body = le_body.transform(daf.email_body)
daf.subject = le_sub.transform(daf.subject)
daf.email_url = le_url.transform(daf.email_url)
daf.head()
```

```python
print(df.head())
print(daf.head())
```

```python
final = pd.merge(df, daf, on='campaign_id')
final.head()
```

```python
final.info()
```

```python
final.to_csv("./data/modified_train.csv", header=True, index=False)
```

```python
import pandas as pd
import numpy as np
import keras
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv("./data/modified_train.csv")
```

```python
X = df.drop(['is_open','is_click','id','user_id','campaign_id'], axis=1)
y = df[['is_open','is_click']]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
print("Scaling the data.")
scaler = StandardScaler()
X_scaled = pd.DataFrame(scaler.fit_transform(X_train), columns=X_train.columns)
X_scaled.head()
```

```python
from keras.models import Sequential
from keras.layers import Dense
from keras import regularizers
```

```python
model = Sequential()
```

```python
model.add(Dense(12,kernel_regularizer=regularizers.l2(0.1), activation="relu", input_shape=(12,)))
model.add(Dense(1, activation='sigmoid'))

model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['acc'])
```

```python
small_model = model.fit(X_scaled, y_train['is_open'], epochs=10, validation_split=0.3, verbose=True)
```
